```python
# Model Building
def search_courses_by_subject(df, subject):
    # Filter dataset based on subject
    filtered_df = df[df['subject'] == subject]

    # Extract relevant columns
    results = filtered_df[['course_title', 'url', 'price', 'num_subscribers']]

    return results
```

```python
search_courses_by_subject(df,'Business Finance')
```
